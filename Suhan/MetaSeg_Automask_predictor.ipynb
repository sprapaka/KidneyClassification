```python
# display original image
#________________
#from IPython.display import Image
#Image("OsteosarcomaMRI.jpeg")
#_________________

from PIL import Image
from IPython.display import display

img1 = Image.open('0001.tiff', 'r')
display(img1)





```

```python
# Import the mask predictor from metaseg
from metaseg import SegAutoMaskPredictor
```

```python
# predict the automask using SegAutoMaskPredictor
#using different model types produces slightly different masks
results = SegAutoMaskPredictor().image_predict(
    source="0001.tiff",
    model_type="vit_b", # vit_l, vit_h, vit_b
    points_per_side=16,
    points_per_batch=64,
    min_area=0,
    output_path="output_b.png",
    show=False,
    save=True,
)
```

```python
results = SegAutoMaskPredictor().image_predict(
    source="0001.tiff",
    model_type="vit_h",
    points_per_side=40,
    points_per_batch=128,
    min_area=0,
    output_path="output_h.png",
    show=False,
    save=True,
)
```
